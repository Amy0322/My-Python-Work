```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

#### 載入資料並檢視資料結構

```python
df1 = pd.read_csv("C:/Users/USER/Desktop/a_lvr_land_a.csv")
df2 = pd.read_csv("C:/Users/USER/Desktop/a_lvr_land_b.csv")
```

```python
df1
```

```python
df2
```

#### 刪除不必要資料並合併

```python
df1 = df1.drop([0])
df2 = df2.drop([0])
```

```python
df = pd.concat([df1,df2],axis=0,ignore_index=True)
```

```python
df_ = df[['鄉鎮市區','土地移轉總面積平方公尺','單價元平方公尺','都市土地使用分區']]
```

```python
df_
```

#### 取得鄉鎮市區各個買賣累積次數

```python
df_['鄉鎮市區'].value_counts()
```

#### 將各鄉鎮市區轉為array

```python
district = df_['鄉鎮市區'].unique()
```

```python
district
```

#### 載入中文字型並完成圓餅圖

```python
from matplotlib import font_manager
my_font = font_manager.FontProperties(fname="C:/Windows/Fonts/msjh.ttc")
labels = district
size = df_['鄉鎮市區'].value_counts()
# patches：為了得到餅圖的返回值
# l_texts：餅圖外標籤的文字
# p_texts：餅圖內部文字的
# autopct：圓裡面的文字格式，％1.1f %%表示小數有1位，整數有1位的浮點數
patches,l_text,p_text = plt.pie(size,labels=labels,autopct='%1.1f%%',textprops={'fontsize':12})
# 把每個文字設成中文字型
for t in l_text: 
    t.set_fontproperties(my_font)
# 保持餅形圖圓形，保證長寬一樣
plt.axis('equal')
plt.title('台北市不動產買賣各區比例圖',fontsize=20,fontproperties=my_font);
```

#### 以相同方式取得都市土地使用分區圓餅圖

```python
df_['都市土地使用分區'].value_counts()
```

```python
use = df_['都市土地使用分區'].unique()
```

```python
use
```

#### 刪除nan值

```python
use = np.delete(use,3,axis=0)
```

```python
use
```

```python
labels = use
size = df_['都市土地使用分區'].value_counts()
# patches：為了得到餅圖的返回值
# l2_texts：餅圖外標籤的文字
# p2_texts：餅圖內部文字的
# autopct：圓裡面的文字格式，％1.1f %%表示小數有1位，整數有1位的浮點數
patches,l2_text,p2_text = plt.pie(size,labels=labels,autopct='%1.1f%%',textprops={'fontsize':12})
# 把每個文字設成中文字型
for t in l2_text: 
    t.set_fontproperties(my_font)
# 保持餅形圖圓形，保證長寬一樣
plt.axis('equal')
plt.title('台北市不動產買賣個土地使用分區比例圖',fontsize=20,fontproperties=my_font);
```
