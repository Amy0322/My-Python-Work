## 二、分析107嬰兒出生數按性別、胎次及生母年齡

```python
df = pd.read_csv("https://quality.data.gov.tw/dq_download_csv.php?nid=100324&md5_url=9a4973172cf2d4d88135f8ac9fe02829")
```

```python
df.columns = ['統計年度','按照別','區域別','胎次','出生者性別','生母年齡','嬰兒出生數']
df
```

### 刪除第0列、按照別行和嬰兒出生數為0的資料

```python
#刪除第0列
df.drop(0,axis=0,inplace=True)
#刪除按照別行
df.drop('按照別',axis=1,inplace=True)
#刪除嬰兒出生數為0的資料
df = df[~df['嬰兒出生數'].str.contains('0')]
```

```python
df
```

### 由大到小依序排列嬰兒出生數

```python
df_sorted = df.sort_values(by=['嬰兒出生數'],ascending=False)
```

```python
df_sorted
```

## 三、分析歷年來台旅客國籍

```python
travel = pd.read_csv('https://quality.data.gov.tw/dq_download_csv.php?nid=45537&md5_url=0081346529af3c5fef2835eaa84c27c8')
```

```python
travel
```

### 刪除不必要資料

```python
travel.drop(index=[9,11,13,20,33,37,40,43],axis=0,inplace=True)
travel.drop(['國籍','細分'],axis=1,inplace=True)
```

```python
travel
```

```python
def convert(value):
    new_value = value.replace(',', '')
    return np.int(new_value)
travel['2003']=travel['2003'].apply(convert)
travel['2004']=travel['2004'].apply(convert)
travel['2005']=travel['2005'].apply(convert)
travel['2006']=travel['2006'].apply(convert)
travel['2007']=travel['2007'].apply(convert)
travel['2008']=travel['2008'].apply(convert)
travel['2009']=travel['2009'].apply(convert)
travel['2010']=travel['2010'].apply(convert)
travel['2011']=travel['2011'].apply(convert)
travel['2012']=travel['2012'].apply(convert)
travel['2013']=travel['2013'].apply(convert)
travel['2014']=travel['2014'].apply(convert)
travel['2015']=travel['2015'].apply(convert)
travel['2016']=travel['2016'].apply(convert)
```

```python
travel
```

### 反轉行列

```python
travel = travel.T
```

```python
travel
```

### 更改columns名稱

```python
travel.columns=['Japan','Korea','India','Middle East','Malaysia','Singapore','Indonesia','Philippines','Thailand','South East Asia Others','Asia Others','Canada','USA','Mexico','Brazil','Argentina','America Others','Belgium','France','Germany','Italy','Netherlands','Switzerland','Spain','UK','Austria','Greece','Sweden','Europe Others','Australia','New Zealand','Oceania Others','S.Africa','Africa Others','Unstated','Overseas Chinese']
```

```python
travel
```

```python
travel.plot();
plt.xlabel('Year')
plt.ylabel("Number")
plt.title("Nationality of visitors who come to Taiwan over the years")
plt.figure(figsize=(200,200))
```
