## 挑戰題一：買相機

```python
m = [1,2,3,4]
list = []
c1 = 100000
c2 = c1+150000
c3 = c2+230000
c4 = c3+80000
list.append(c1)
list.append(c2)
list.append(c3)
list.append(c4)
print('折線圖')
plt.plot(m,list);
```

```python
print('圓餅圖')
months =["1","2","3","4"]
class_a = [100000/c4,150000/c4,230000/c4,80000/c4]
plt.pie(class_a,labels=months);
```
