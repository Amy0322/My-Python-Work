```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandas_datareader.data as web
```

### 1.使用pandas_datareader得到 "AAPL" 這間公司股價的所有資料，不用設定資料時間

```python
df = web.DataReader('AAPL',"yahoo")
```

### 2.取出最後300筆資料

```python
data = df.tail(300)
```

```python
data
```

### 3.從這300筆資料，取出收盤價close，並轉成numpy_array

```python
close = data['Close'].values
```

### 4.用前一天的收盤價，預測第二天的收盤價（星期一預測星期二），把資料分成題目X(表示前一天收盤價)，跟答案Y(第二天收盤價) 

```python
X = close[:-1]
Y = close[1:]
```

### 5.把題目X，答案Y再分成佔80%的訓練資料跟佔20%的測試資料

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=85)
```

```python
x_train_len = len(x_train)
x_train.shape = (x_train_len,1)
```

### 6.利用訓練資料和scikit learn跑線性回歸，畫圖，並在得到的迴歸直線上，加上訓練資料的資料點

```python
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(x_train,y_train)
y_train_predict = regr.predict(x_train)

plt.scatter(x_train,y_train)
plt.plot(x_train,y_train_predict,'r')
plt.xlabel('Previous')
plt.ylabel('Future')
```

### 7.回歸的結果跟實際結果差多少

```python
x_test_len = len(x_test)
x_test.shape = (x_test_len,1)

y_test_predict = regr.predict(x_test)

plt.scatter(x_test,y_test)
plt.plot(x_test,y_test_predict,'r')
plt.xlabel('Previous')
plt.ylabel('Next')
```
