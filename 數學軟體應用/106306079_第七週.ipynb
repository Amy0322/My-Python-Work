```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandas_datareader.data as web
```

### 1.使用pandas套件讀檔

```python
df = pd.read_csv('train.csv')
df
```

### 2.取出所有的Age跟Sex作為題目，和Survived做為答案並分成三個同名的numpy array

```python
age = df['Age'].values
sex = df['Sex'].values
survived = df['Survived'].values
```

### 3.其中，由於Age有一些值是空值NaN，請把這些空值都先換成0

```python
# isnan:得到NaN值
age[np.isnan(age)] = 0
age = age.reshape(891,1)
df
```

### 4.Sex部分因為都是文字，不方便我們做svm，請把Male都換成0，Female都換成1

```python
sex[sex == 'male'] = 0
sex[sex == 'female'] = 1
sex = sex.reshape(891,1)
df
```

### 5.把Sex和Age兩個array合併成一個

```python
sex_age = np.concatenate([sex,age],axis = 1)
sex_age
```

### 6.拿svm 訓練看看，並用同樣的data，predict看看結果

```python
from sklearn.svm import SVC
```

```python
clf = SVC(gamma='auto',random_state=87)
```

```python
clf.fit(sex_age,survived)
```

```python
Y_predicted = clf.predict(sex_age)
plt.scatter(sex_age[:,0],sex_age[:,1],c=Y_predicted)
plt.xlabel('Sex')
plt.ylabel('Age')
print('預測結果')
```

```python
plt.scatter(sex_age[:,0],sex_age[:,1],c=Y_predicted-survived)
plt.xlabel('Sex')
plt.ylabel('Age')
print('預測與實際的差距')
```
