```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 成績分布直方圖

```python
x = np.random.randint(0,101,40)
count = 0
for i in range(len(x)):
    if(x[i]>=60):
        count+=1
print('最高分:',np.max(x),'\n最低分:',np.min(x),'\n及格人數:',count,'\n平均數:',np.mean(x),'\n標準差:',np.std(x))
plt.hist(x,bins=10,range=[0,100],rwidth=0.95)
plt.xlabel('score')
plt.ylabel('number')
```

## 改變散佈圖點的形狀

```python
P = np.random.randn(50,2)
a = int(input("請問希望散佈圖點的形狀為?（請輸入數字）\n1.圓點\n2.三角形\n3.正方形\n4.八角形\n5.星號\n"))
if a==1:
    plt.scatter(P[:,0],P[:,1],marker='o')
elif a==2:
    plt.scatter(P[:,0],P[:,1],marker='^')
elif a==3:
    plt.scatter(P[:,0],P[:,1],marker='s')
elif a==4:
    plt.scatter(P[:,0],P[:,1],marker='8')
elif a==5:
    plt.scatter(P[:,0],P[:,1],marker='*')
```

## 將符合標準的標示出來

```python
c=np.random.randint(1,101,30)
m=np.random.randint(1,101,30)
c_passed=c[(c>=60)]
m_passed=m[(m>=60)]
plt.scatter(c,m)
plt.scatter(c_passed,m_passed,c='r')
```
